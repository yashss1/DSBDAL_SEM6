```python
import pandas as pd
```

```python
DataFrame=pd.read_csv('movies_metadata.csv',low_memory=False)
DataFrame
```

# Subset

```python
Subset1=DataFrame[DataFrame['original_language']=='en']
Subset1
```

```python
Subset2=DataFrame[DataFrame['original_language']=='fa']
Subset2
```

```python
Subset3=DataFrame[DataFrame['original_language']=='tl']
Subset3
```

```python
Subset4=DataFrame[DataFrame['original_language']=='zh']
Subset4
```

```python
Subset5=DataFrame[DataFrame['original_language']=='it']
Subset5
```

```python
Subset6=DataFrame[DataFrame['original_language']=='nl']
Subset6
```

```python
Subset7=DataFrame[DataFrame['original_language']=='da']
Subset7
```

# Merge

```python
Merged1=pd.concat([Subset1,Subset2])
Merged1
```

```python
Merged2=pd.concat([Subset3,Subset4])
Merged2
```

```python
Merged3=pd.concat([Subset5,Subset6,Subset7])
Merged3
```

# Transpose

```python
DataFrame.T
```

# Melt

```python
Melting=pd.melt(DataFrame,id_vars=['id','original_language'],var_name='Metric',value_name='value')
Melting
```

# Casting

```python
Casting=pd.pivot_table(DataFrame,index=['id','original_language'],values=['original_title','spoken_languages'])
Casting
```

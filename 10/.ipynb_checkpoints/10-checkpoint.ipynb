```python
import pandas as pd
```

```python
DataFrame=pd.read_csv('Iris.csv')
DataFrame
```

# Subset

```python
Subset1=DataFrame[DataFrame['class']=='Iris-setosa']
Subset1
```

```python
Subset2=DataFrame[DataFrame['class']=='Iris-virginica']
Subset2
```

```python
Subset3=DataFrame[DataFrame['class']=='Iris-versicolor']
Subset3
```

# Merge

```python
Merged=pd.concat([Subset1,Subset2])
Merged
```

# Sort

```python
Sorted=DataFrame.sort_values('petal length in cm')
Sorted
```

# Transpose

```python
Transposed=DataFrame.T
Transposed
```

# Melting

```python
Melting=pd.melt(DataFrame,id_vars=['sepal length in cm','petal length in cm'],value_name='value',var_name='Metrics')
Melting
```

# Casting

```python
Casting=pd.pivot_table(DataFrame,index=['sepal length in cm','petal length in cm'],values=['sepal width in cm','petal width in cm'])
Casting
```

```python
import pandas as pd
```

```python
df=pd.read_csv('Iris.csv')
df.columns =["sepal length in cm","sepal width in cm","petal length in cm","petal width in cm","class"]
df.head()
```

```python
df['class'].unique()
```

# Subset

```python
df1=df[df['class']=='Iris-setosa']
df1.head()
```

```python
df2=df[df['class']=='Iris-virginica']
df2.head()
```

```python
df3=df[df['class']=='Iris-versicolor']
df3.head()
```

# Merge

```python
df4 = pd.concat([df1,df2])
df4
```

# Sort

```python
df5=df.sort_values('petal length in cm',ascending=True)
df5
```

# Transpose

```python
df6 = df.T
df6
```

# Melting

```python
Melting = pd.melt(df,id_vars=['sepal length in cm','petal length in cm'],value_name='value',var_name='Metrics')
Melting
```

# Casting

```python
Casting=pd.pivot_table(df,index=['sepal length in cm','petal length in cm'],values=['sepal width in cm','petal width in cm'])
pd.set_option("display.max_rows", None)
Casting

```

```python
import pandas as pd
import matplotlib.pyplot as mat
import seaborn as sns
```

```python
DataFrame=pd.read_csv('heart.csv')
DataFrame
```

# HIST

```python
col=['trestbps','chol','thalach']
for column in col:
    DataFrame[column].hist(bins=10,figsize=(10,6))
    mat.tight_layout()
    mat.show()
```

# PIE 

```python
gender_labels = ['Male', 'Female']
gender_counts = [713,312]

mat.figure(figsize=(6, 6))
mat.pie(gender_counts, labels=gender_labels, autopct='%1.1f%%')
mat.title('Gender Distribution')
mat.show()
```

# BOX

```python
mat.figure(figsize=(6, 6))
sns.boxplot(data=DataFrame, x='target', y='trestbps')
mat.title('Box Plot - Resting Blood Pressure')
mat.xlabel('Target')
mat.ylabel('Resting Blood Pressure')
mat.show()
```

# Scatter Plot

```python
col=['chol','thalach']
sns.scatterplot(data=DataFrame,x=col[0],y=col[1],hue='target')
mat.xlabel(col[0])
mat.ylabel(col[1])
mat.show()
```

# Bar

```python
col=['restecg','fbs','slope']
for column in col:
    DataFrame[column].value_counts().plot(kind='bar')
    mat.show()
```

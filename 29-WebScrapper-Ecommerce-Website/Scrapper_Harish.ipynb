```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

```python
reviews = soup.findAll('div',{'class' : 't-ZTKy'})
for review in reviews:
    print(review.get_text()+"\n\n")
```

```python
# Average rating
ratings = soup.find('div',{'class':'_2d4LTz'}).get_text()
print(ratings)
```

```python
# Individual rating
in_ratings = soup.findAll('div',{'class':'_3LWZlK _1BLPMq'})
for rating in in_ratings:
    print(rating.get_text()+"\n")
```

```python
# Customer Names
cust_name = soup.findAll('p',{'class':'_2sc7ZR _2V5EHH'})
for customer in cust_name:
    print(customer.get_text() +"\n")
```

```python
#fetch price of product
price = soup.find('div',{'class' :'_30jeq3 _16Jk6d'}).get_text()
print(price)
```

```python
questions = soup.findAll('div',{'class':'_1xR0kG _3cziW5'})
for question in questions:
    print(question.get_text())
    answer = question.findNext('div',{'class':'_2yeNfb'}).get_text()
    print(answer +"\n\n")
```

```python
# Specifications
specifications = soup.findAll('td',{'class':'_1hKmbr'})
for specification in specifications:
    print(specification.get_text())
    answer = specification.findNext('li',{'class':'_21lJbe'}).get_text()
    print(answer+"\n\n")
```

```python
import pandas as pd
import matplotlib.pyplot as mat
import seaborn as sn
```

```python
DataFrame=pd.read_csv('heart.csv')
DataFrame
```

```python
DataFrame.isna().sum()
```

# HIST

```python
col=['age','chol','thalach']
DataFrame[col].hist(bins=10,figsize=(10,6))
mat.tight_layout()
mat.show()
```

# DOT

```python
col=['sex', 'cp', 'target']
for column in col:
    sn.catplot(x=column,kind='count',data=DataFrame,aspect=1.5)
mat.plot()
```

# BAR 

```python
col=['restecg','fbs','slope']
for column in col:
    DataFrame[column].value_counts().plot(kind='bar')
    mat.xlabel(col)
    mat.ylabel('Count')
    mat.show()
```

# LINE

```python
columns = ['trestbps', 'oldpeak', 'thalach']
for column in columns:
    mat.plot(DataFrame[column])
    mat.xlabel('Index')
    mat.ylabel(column)
    mat.show()
```

# Multivariate analysis using scatter plot

```python
# Plot scatter plot for two selected columns
columns = ['chol', 'thalach']
sn.scatterplot(data=DataFrame, x=columns[0], y=columns[1], hue='target')
mat.xlabel(columns[0])
mat.ylabel(columns[1])
mat.show()
```
